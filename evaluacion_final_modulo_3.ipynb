CANCELLATION YEAR & CANCELLATION MONTH NULOS

```python
df_all["Cancellation Year"].value_counts() #cancelaciones por año
```

```python
df_all["Cancellation Month"].value_counts() #cancelaciones por me
```

NOTA: 
* Dejamos los valores nulos de Cancelation year y month tal como estan , ya que el nulo indica que el cliente no ha cancelado y esta informacion puede llegar a ser util.

```python
# Voy hacer una funcion para clasificar el esatdos
def get_status(row):# row es cada fila del df
    if pd.notnull(row["Cancellation Year"]) or pd.notnull(row["Cancellation Month"]):
        return "inactivo"
    else:
         return "Activo"
df_all["status"]=df_all.apply(get_status,axis= 1) # crea una columna nueva lllmada status

# si cancellation year o month tienen un dato marcalo como inactivo ( quiere decir que cancelaron en una fecha)
# si no tiene dato pon activo , (no han cancelado )

```

```python
df_all.head() # confirmo que exista la columna Status
```

NOTA:
* En lugar de modificar los datos nulos de Cancellation year y month he decidido dejarlos en su forma original , de esta forma conservo la informacion original .
* Cree una columna status que clasifica en activo o inactivo, si existe l apresencia de una fecha de cancelacion. Esta informacion puede ser util para el futuro
